```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cpu = np.loadtxt("plot/times-CPU.dat").T
gpu = np.loadtxt("plot/times-GPU.dat").T
```

```python
nc, tc1, tc2, tc3 = cpu
ng, tg1, tg2, tg3 = gpu
```

```python
plt.figure(figsize=(12, 8))
plt.plot(nc, tc2, 'o-', lw=3, label='CPU')
plt.plot(ng, tg2, 'o-', lw=3, label='GPU')
plt.legend(fontsize=24)
plt.xlabel('Matrix Size', fontsize=24)
plt.ylabel('Time (s) in Diagonalization', fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
# plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(nc, tc1, 'o-', lw=3, label='CPU')
plt.plot(ng, tg1, 'o-', lw=3, label='GPU')
plt.legend(fontsize=24)
plt.xlabel('Matrix Size', fontsize=24)
plt.ylabel('Time (s) in Matrix Generation', fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
# plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(nc, tc3, 'o-', lw=3, label='CPU')
plt.plot(ng, tg3, 'o-', lw=3, label='GPU')
plt.legend(fontsize=24)
plt.xlabel('Matrix Size', fontsize=24)
plt.ylabel('Time (s) in Matrix Mult', fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
# plt.yscale('log')
plt.show()
```
